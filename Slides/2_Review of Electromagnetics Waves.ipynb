# Review of Electromagnetic Waves

## Notations

* Vector will be underlined ($\underline{E}$). 
* Tensor will be double underlined ($\underline{\underline{K}}$)

## Maxwell Equations
### Macroscopic Equations
Starting from the Maxwell equations is their most general form, before recasting them to fit our needs:

$$ \nabla \times \underline{E} = - \frac{\partial \underline{B}}{\partial t}$$

$$ \nabla \times \underline{H} = \frac{\partial \underline{D}}{\partial t} + \underline{J} $$

$$ \nabla \cdot \underline{D} = \rho $$

$$ \nabla \cdot \underline{B} = 0 $$

Where the usual electromagnetic field quantities are:
* $\underline{E}$ Electric Field Intensity (in $V/m$) 
* $\underline{D}$ Electric Flux Density (in $A.s/m^2 = C/m^2$)
* $\underline{H}$ Magnetic Field Intensity (in $A/m$)
* $\underline{B}$ Magnetic Flux Density (in $V.s/m^2 = Wb/m^2 = T$)
* $\underline{J}$ (Total) Current Density (in $A/m^2$)
* $\rho$ (Total) Electric Charge Density (in $C/m^3$)



### Notes
* All quantities are defined in space ($\underline{r}$) and time ($t$), i.e. $\underline{E}=\underline{E}(\underline{r}, t)$.
* In a fully ionized plasma, total current and charge density can be decomposed into:
  * *free*: carried by free charges flowing into the medium (plasma) 
  * *ext*: generated by external sources (such as antennas)
  $$ \underline{J} = \underline{J_{free}} + \underline{J_{ext}}$$
  $$ \rho = \rho_{free} + \rho_{ext} $$
  
* Charge conservation can be viewed as a consequence of Maxwell's Equations:
$$
\frac{\partial \rho}{\partial t} + \nabla \cdot \underline{J} = 0 
$$
* Maxwell-Gauss laws can be seen as Initial Conditions
* Maxwell-Faraday and Maxwell-Amperes laws can be seen as evolution equations

## Phase and Group Velocity

### phase velocity
Let's assumed the $phase$ of a wave is $\theta = k x - \omega t$

The velocity of the phase moves with the crest of a wave. The rate of change of the phase doesn't change: $d\theta/dt=0$  so 
$$ x \frac{d x}{d t} - \omega \frac{d t}{d t} = 0$$
So
$$k v_\phi - \omega = 0$$
and finally the *phase velocity* 
$$ v_\phi = \frac{\omega}{k}$$

### Group Velocity

Group velocity is the velocity of the envelope, defined as
$$ v_g = \frac{d \omega}{d k} $$

## Constitutive Relations
### Generic Expressions
* Flux Densities $(\underline{D}, \underline{B})$ differ from field intensities  $(\underline{E}, \underline{H})$ inside the material (magnitude and direction)
    * Flux Densities can be interpreted as a response of the medium to an applied excitation
* To be able to solve the previous four equations, charge and current must be known.
* One needs to specify the relationships existing between flux densities $(\underline{D}, \underline{B})$ and electric current density $(\underline{J})$ with intensities $(\underline{E}, \underline{H})$, these are the *constitutive relations*:
$$ \underline{D} = \underline{D}(\underline{E}, \underline{H}) $$
$$ \underline{B} = \underline{B}(\underline{E}, \underline{H}) $$
$$ \underline{J} = \underline{J}(\underline{E}, \underline{H}) $$

* For real-world materials, constitutive relations are rarely simple, except approximately, and usually determined by experimentally

### Polarization and Magnetization
In classical electromagnetics, constitutive relationship usually involves polarization ($\underline{P}$) and magnetization  ($\underline{M}$) vectors defined as:

$$  \underline{D} = \varepsilon_0 \underline{E} + \underline{P}$$
$$  \underline{B} = \mu_0 \underline{H} + \mu_0 \underline{M}$$

where $\underline{P}$ and $\underline{M}$ are defined in terms of microscopic bound charges and bound currents.

Using these definitions in Maxwell-Ampere law leads to:
$$
\nabla \times \underline{B}
= 
\mu_0 ( \underline{J_{free}} + \underline{J_{ext}} + \underline{J_{pol}} + \underline{J_{mag}} )
+ \frac{1}{c^2} \frac{\partial \underline{E}}{\partial t}
$$
with:
* $\underline{J_{pol}} = \frac{\partial \underline{P}}{\partial t}$ the polarization current
* $\underline{J_{mag}} = \mu_0^{-1}(\nabla \times \underline{M}) $ the magnetization current

In plasmas, it is impractical to separate polarization, magnetization and free charges currents, since all charges are free (in fully ionized plasma). So all currents contribute to the polarization current of the medium and we rewrite:
$$
\nabla \times \underline{B}
= 
\mu_0 ( \underline{J} + \underline{J_{ext}} )
+ \frac{1}{c^2} \frac{\partial \underline{E}}{\partial t}
$$
where $\underline{J}$ is the *total* current flowing in the plasma in response to the wave perturbation. 

### Wave Equation in Plasmas
The wave equation can be deduced:
$$
\nabla \times \nabla \times \underline{E} + \frac{1}{c^2} \frac{\partial^2 \underline{E} }{\partial t^2}
= 
-\mu_0 \frac{\partial}{\partial t} \left( \underline{J} + \underline{J_{ext}} \right)
$$
which is non-linear by nature as $\underline{J}$ depends on $\underline{E}$ from the constitutive relation of the medium.

* Establishing a *plasma model* consists in defining the *constitutive relation* $\underline{J}(\underline{E})$
* Two main classes of plasma model:
    * Fluid models: deal with averaged quantities rather thab with the distribution functions themselves.
    * Kinetic models
* NB: a kinetic model is needed to describe non-collisional damping processes.

## Properties of Magnetized Plasma
### Anisotropy
In addition, response can depends of the direction of excitation: *anisotropy* (tensorial relations):

$$
\underline{J}(\underline{r}, t) = \underline{\underline{\sigma}} \cdot \underline{E}(\underline{r},t)
$$

### Dispersion
* If a medium exhibits a time or space dependence to an electromagnetic excitation, it is said to be *dispersive*. 
* Its phase velocity depends on the frequency. 

 <img src="figures/Optical_dispersion_dynamics.gif" alt="https://en.wikipedia.org/wiki/Dispersion_(optics)#/media/File:Optical_dispersion_dynamics.gif" />

#### Time-dispersion
* Depending on wave frequency, time delay between the imposition of the electric field and the resulting polarization of the medium
* Due to the inertia of charged particles (various weights) to respond to the time-varying field (ions slower than electrons)
$$ \underline{J}(\underline{r}, t)
=
\int_{\infty}^t dt' \underline{\underline{\sigma}}(\underline{r}, t, t') \cdot \underline{E}(\underline{r}, t')
$$

#### Space-dispersion

* Response at the location $r$ and time $t$ not only depends on the field at ($r$, $t$),
    * but of the field in its vicinity and by all previous instant. 

$$ \underline{J}(\underline{r}, t)
=
\int d^3 \underline{r}' \underline{\underline{\sigma}}(\underline{r}, \underline{r}', t') \cdot \underline{E}(\underline{r}',t)
$$

* In plasma, thermal agitation of the species induces an additional erratic motion to the particles trajectory.
    * Particles are influenced by the field in the domain explored by their motion.
    * This space dispersion can be omitted in the limit at which the temperature effects can be neglected.    

### General Case
Hence, combining the 3 lattest properties, the general case is:

$$
\underline{J} (\underline{r}, t) 
=  
\int_{t'=-\infty}^{t} d t'
\int d \underline{r}' \;
\underline{\underline{\sigma}}  (\underline{r},\underline{r}', t,t') \cdot \underline{E} (\underline{r}', t') 
$$


## Harmonic Regime
### Phasors
RF source time excitation can be approximated by varying sinusoidally in time with a single frequency $\omega_0$. This is called *harmonic regime*:
$$
a(\underline{r},t) = |A(\underline{r})| \cos\left(\omega_0 t + \phi_0 \right) = \Re \left[A(\underline{r}) e^{-i\omega_0 t} \right]
$$
with $A$ a complex quantity, or *phasor*: $A(\underline{r})=|A(\underline{r})|e^{i\phi_0}$, which only depend of $\underline{r}$.

This relation can be extended to vector quantities:
$$
\underline{\mathcal{E}}(\underline{r},t) = \Re \left[\underline{E}(\underline{r}) e^{-i\omega_0 t} \right]
$$

**Properties**
* Time-average (RMS):
$$
\left< \underline{\mathcal{E}}(t) \right>
=
\left[ \frac{1}{T}\int_0^T dt |\underline{\mathcal{E}}(t)|^2 \right]^{1/2}
=
\frac{|\underline{E}|}{\sqrt{2}} 
$$

* Linearity:
$$
\nabla \cdot \mathcal{\underline{B}}(\underline{r},t) = 0  \Longrightarrow \nabla \cdot \underline{B}(\underline{r}) = 0
$$

### $k-\omega$ Representation
Let us generalise to the case where the field solution is represented by the summation of many plane (or evanescent) waves characterized by their wavevector $k = (k_x; k_y; k_z)$. We construct a solution on the form:
$$
\underline{E}(\underline{r},t)
=
\Re\left[
\int d\underline{k}
\int d\omega
\underline{E}(\underline{k},\omega)
e^{i(\underline{k}\cdot\underline{r} - \omega t)}
\right]
$$
since $\underline{E}(\underline{r},t)$ is a real (physical) quantity, the real part can be dropped.

We have the inverse transform 
$$
\underline{E}(\underline{k},\omega)
=
\frac{1}{(2\pi)^4}
\int d\underline{r}
\int d t
\underline{E}(\underline{r},t)
e^{-i(\underline{k}\cdot\underline{r} - \omega t)}
$$

**NB**: the case for a single frequency $\omega_0$ corresponds to:

$$\underline{E}(r, \omega) = \underline{E}_0(r) \delta(\omega-\omega_0)$$

and for a single monochromatic plane-wave is

$$\underline{E}(r, \omega) = \underline{E}_0(r) \delta(\omega-\omega_0)\delta(\underline{k}-\underline{k}_0)$$

## Maxwell's Equation for Linear Plasma Waves
### Stationary and Homogeneous Plasmas
If insert the previous constitutive relation for the current density:
$$
\underline{J} (\underline{r}, t) 
=  
\int_{t'=-\infty}^{t} d t'
\int d \underline{r}' \;
\underline{\underline{\sigma}}  (\underline{r},\underline{r}', t,t') \cdot \underline{E} (\underline{r}', t') 
$$

In the wave equation:
$$
\nabla \times \nabla \times \underline{E} + \frac{1}{c^2} \frac{\partial^2 \underline{E} }{\partial t^2}
= 
-\mu_0 \frac{\partial}{\partial t} \left( \underline{J} + \underline{J_{ext}} \right)
$$


we end up with an integro-differential equation, which is linear but very difficult to solve.

To make further progress in the physical understanding of the waves in plasma we need to make additional assumptions.

#### Homogeneous Plasma
Medium is invariant by translation: Conductivity can only depend on the difference between excitation location $\underline{r}'$ and the response location $\underline{r}$:
$$
\underline{\underline{\sigma}}
(\underline{r}, \underline{r}', t, t') = 
\underline{\underline{\sigma}}
(\underline{r} - \underline{r}', t, t')
$$

#### Stationary Plasma
Medium is invariant in time: Conductivity can only depend on the difference between excitation time $t'$ and the response time $t$:
$$
\underline{\underline{\sigma}}
(\underline{r}, \underline{r}', t, t') = 
\underline{\underline{\sigma}}
(\underline{r}, \underline{r}', t- t')
$$


#### Homogeneous and Stationary Plasma
Using both assumptions, we have:
$$
\underline{J} (\underline{r}, t) 
=  
\int_{t'=-\infty}^{t} d t'
\int d \underline{r}' \;
\underline{\underline{\sigma}}  (\underline{r}-\underline{r}', t-t') \cdot \underline{E} (\underline{r}', t') 
$$
which leads in the $(k-\omega)$ space to the simple relationship:

$$
\underline{J}(\underline{k},\omega)
=
\underline{\underline{\sigma}}(\underline{k},\omega)
\cdot
\underline{E}(\underline{k},\omega)
$$

### Wave Equation

The wave equation in the $(k-\omega)$ space now reads:
$$
\underline{k} \times \underline{k} \times \underline{E}(\underline{k},\omega)
+ \frac{\omega^2}{c^2} \underline{E}(\underline{k},\omega)
= 
 i \omega \mu_0 
\underline{\underline{\sigma}}(\underline{k},\omega)
\cdot
\underline{E}(\underline{k},\omega)
+ i \omega \mu_0 \underline{J_{ext}}(\underline{k},\omega)
$$

$$
\underline{k} \times \underline{k} \times \underline{E}(\underline{k},\omega)
+ 
\frac{\omega^2}{c^2} 
\left(
\underline{\underline{1}}
+
\frac{i}{\omega \varepsilon_0}  
\underline{\underline{\sigma}}(\underline{k},\omega)
\right) \cdot
\underline{E}(\underline{k},\omega)
= 
+ i \omega \mu_0 \underline{J_{ext}}(\underline{k},\omega)
$$


Or, finally:
$$
\underline{k} \times \underline{k} \times \underline{E}(\underline{k},\omega)
+ 
\frac{\omega^2}{c^2} 
\underline{\underline{K}}
\cdot
\underline{E}(\underline{k},\omega)
= 
 i \omega \mu_0 \underline{J_{ext}}(\underline{k},\omega)
$$
with the normalized dielectric tensor $\underline{\underline{K}}=\underline{\underline{\varepsilon_r}}/\varepsilon_0$ defined by:
$$
K
=
\underline{\underline{1}}
+
\frac{i}{\omega \varepsilon_0}  
\underline{\underline{\sigma}}(\underline{k},\omega)
$$
